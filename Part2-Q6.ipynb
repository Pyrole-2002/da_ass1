```python
# Core libraries
import pandas as pd
import numpy as np

# Visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: Set Seaborn style for cleaner plots
sns.set(style='whitegrid')

%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd

# Load the cleaned dataset
df = pd.read_csv('C:/Users/bvgpo/Desktop/IIIT_HYDERABAD/Sem2/DA/A1/cleaned_real_estate_data.csv')
```

```python
mumbai_df = df[df['city'].str.lower() == 'mumbai']
thane_df = df[df['city'].str.lower() == 'thane']


possession_impact = df[df['city'].isin(['Mumbai', 'Thane'])].groupby(['city', 'possession_status'])['price_lakhs'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.boxplot(x='possession_status', y='price_lakhs', hue='city', data=df[df['city'].isin(['Mumbai', 'Thane'])], palette='Set2')
plt.title('Price Distribution by Possession Status: Mumbai vs Thane')
plt.xlabel('Possession Status')
plt.ylabel('Price (Lakhs)')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()

```

```python
#additional insights

#ready to move vs under construction price comparision

possession_price = df[df['city'].isin(['Mumbai', 'Thane'])].groupby(['city', 'possession_status'])['price_lakhs'].mean().reset_index()

plt.figure(figsize=(10, 5))
sns.barplot(x='possession_status', y='price_lakhs', hue='city', data=possession_price, palette='Set2')
plt.title('Avg Price by Possession Status: Mumbai vs Thane')
plt.ylabel('Price (Lakhs)')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
#price spread for under-construction listings

uc_df = df[(df['possession_status'] == 'Under Construction') & (df['city'].isin(['Mumbai', 'Thane']))]

plt.figure(figsize=(8, 5))
sns.boxplot(x='city', y='price_lakhs', data=uc_df, palette='coolwarm')
plt.title('Price Spread for Under-Construction Properties')
plt.ylabel('Price (Lakhs)')
plt.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```
